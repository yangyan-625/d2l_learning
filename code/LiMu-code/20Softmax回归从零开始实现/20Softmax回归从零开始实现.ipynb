```python
import torch
from IPython import display
from d2l import torch as d2l

batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

```python
"""
将展平每个图像，将它们视为长度为784的向量，即有784个像素特征，神经网络的输入维度为784
由于是多类任务，且标签有10个类别，所以神经网络的输出维度为10
"""
num_inputs = 784
num_outputs = 10

"""
题外话
线性回归模型y = Xw + b，X是n*m维，n是样本数，m是特征数，故w是m*1维向量

多类模型softmax(XW + b)：
X是形状为[n, m]的特征矩阵
W是形状为[m, k]的权重矩阵，m是特征数，k是标签的类型总数
b是形状为[k]的偏置向量，与XW相加会使用广播机制
"""
W = torch.normal(0, 0.01, size=(num_inputs, num_outputs), requires_grad=True)
b = torch.zeros(num_outputs, requires_grad=True)
```
